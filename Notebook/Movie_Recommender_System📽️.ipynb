```python
movies_df = movies_df.merge(credits_df, on='title')
```

```python
# Checking again for Null Values
movies_df.isnull().sum()
```

```python
import ast
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies_df['genres'] = movies_df['genres'].apply(convert)
movies_df.head()
```

```python
movies_df['keywords'] = movies_df['keywords'].apply(convert)
movies_df.head()
```

```python
movies_df['cast'] = movies_df['cast'].apply(convert)
movies_df.head()
```

```python
# Keep only the first three elements in the 'cast' column
movies_df['cast'] = movies_df['cast'].apply(lambda x: x[0:3])
```

```python
def fetch_director(text):
    # Initialize an empty list to store director names
    L = []

    # Convert the input text to a Python object using ast.literal_eval
    for i in ast.literal_eval(text):
        # Check if the 'job' key in the current dictionary is 'Director'
        if i['job'] == 'Director':
            # Append the 'name' value to the list
            L.append(i['name'])

    # Return the list of director names
    return L
```

```python
movies_df['crew'] = movies_df['crew'].apply(fetch_director)
```

```python
movies_df.sample(5)
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies_df['cast'] = movies_df['cast'].apply(collapse)
movies_df['crew'] = movies_df['crew'].apply(collapse)
movies_df['genres'] = movies_df['genres'].apply(collapse)
movies_df['keywords'] = movies_df['keywords'].apply(collapse)
```

```python
movies_df.head()
```

```python
movies_df['overview'] = movies_df['overview'].apply(lambda x:x.split())
```

```python
movies_df['tags'] = movies_df['overview'] + movies_df['genres'] + movies_df['keywords'] + movies_df['cast'] + movies_df['crew']
```

```python
new = movies_df.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
```

```python
recommend('Gandhi')
```
